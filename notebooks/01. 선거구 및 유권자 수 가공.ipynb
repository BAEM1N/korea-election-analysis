## 선거인 명부 확정 상황
  - 중앙선거관리위원회 선거통계시스템에서 다운로드 가능
  > http://info.nec.go.kr/main/showDocument.xhtml?electionId=0020200415&topMenuId=BI&secondMenuId=BIPB02

```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
base_path = '../data/voter'
SIDO = os.listdir(base_path)
print(SIDO)
```

```python
# 파일 오픈
df = pd.read_excel(os.path.join(base_path, '[제21대_국회의원선거]_선거인명부_확정상황_서울특별시.xlsx'))
df.head()
```

```python
# 파일 재 오픈
df = pd.read_excel(os.path.join(base_path, '[제21대_국회의원선거]_선거인명부_확정상황_서울특별시.xlsx'), skiprows=2)
df.head()
```

```python
# 시도명 추출
sido = df.columns[0].split('[')[-1].replace(']','')

# 컬럼 재설정
df = df.rename(columns=df.iloc[0])
```

```python
# 선거구데이터만 추출

df = df[df['선거구명'].notnull()]
df = df[df['구시군명'] == '소계']
df.head()
```

```python
# 선거구 명 및 선거구 수 확인
df['선거구명'].unique(), len(df['선거구명'])
```

```python
# 필요 컬럼만 추출
district = df[['선거구명', '인구수\n(선거인명부작성\n기준일현재)']]

# 컬럼명 변경
district.columns = ['선거구명', '유권자 수']

# 유권자 수 컬럼 처리
# 유권자 수 컬럼에 존재하는 ,(콤마)를 제거 후 \n(줄바꿈 기호)를 기준으로 분할하여 첫번째 분할된 레코드만 가지고 옴
district['유권자 수'] = district['유권자 수'].str.replace(',', '').str.split('\n').str[0].astype('int64')

district.head()
```

```python
# 모든 광역시도를 하나의 데이터 프레임으로 결합

election_district = pd.DataFrame()

for file in SIDO:
    df = pd.read_excel(os.path.join(base_path, file), skiprows=2)

    # 시도명 추출
    sido = df.columns[0].split('[')[-1].replace(']','')

    # 컬럼 재설정
    df = df.rename(columns=df.iloc[0])

    df = df[df['선거구명'].notnull()]
    df = df[df['구시군명'] == '소계']

    district = df[['선거구명', '인구수\n(선거인명부작성\n기준일현재)']]
    district.columns = ['선거구명', '유권자 수']
    district['유권자 수'] = district['유권자 수'].str.replace(',', '').str.split('\n').str[0].astype('int64')
    district['광역시도명'] = sido

    if sido[-2:] in ['남도', '북도']:
        district['시도명'] = sido[0] + sido[2]
    else:
        district['시도명'] = sido[:2]


    election_district = pd.concat([election_district, district])
    election_district.reset_index(drop=True, inplace=True)
    
election_district.head()
```

```python
election_district.sort_values('선거구명')
```

```python
# 광역시도별 선거구 수 확인

election_district.groupby('광역시도명')['선거구명'].agg(['count']).sort_values('count', ascending=False)
```

```python
# 선거구 수 확인
election_district.groupby('광역시도명')['선거구명'].agg(['count']).sum()
```

```python
# 광역시도별 유권자수 수 확인
election_district.groupby('광역시도명')['유권자 수'].agg(['sum']).sort_values('sum', ascending=False)
```

```python
# 유권자 수 대비 선거구 수 확인
district = election_district.groupby('광역시도명')['선거구명'].agg(['count'])
voter = election_district.groupby('광역시도명')['유권자 수'].agg(['sum'])

rate = voter['sum'] / district['count']
rate.sort_values(ascending=False)
```
