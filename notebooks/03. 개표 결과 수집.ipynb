```python
def make_info(sido, district):
    driver.find_element_by_id("cityCode").send_keys(sido)
    time.sleep(5)
    driver.find_element_by_id("sggCityCode").send_keys(district)
    time.sleep(5)
    element = driver.find_element_by_xpath("""//*[@id="townCodeFromSgg"]""")
    options = element.find_elements_by_tag_name("option")
    for option in options[1:]:
        election_district[sido][district][option.text] = {}
    print(election_district[sido][district].keys())
```

```python
for city in list(election_district.keys()):
    for dist in list(election_district[city].keys()):
        make_info(city, dist)
```

```python
req = []
for sido in election_district.keys():
    for ssg_sido in election_district[sido].keys():
        for ssg_gusigun in election_district[sido][ssg_sido].keys():
            req.append([sido, ssg_sido, ssg_gusigun])
print(req, len(req))
```

```python
def get_data(sido, ssg_sido, ssg_gusingu):
    driver.find_element_by_id("cityCode").send_keys(sido)
    time.sleep(2)
    driver.find_element_by_id("sggCityCode").send_keys(ssg_sido)
    time.sleep(2)
    driver.find_element_by_id("townCodeFromSgg").send_keys(ssg_gusingu)
    time.sleep(2)
    driver.find_element_by_xpath("""//*[@id="spanSubmit"]/input""").click()
    html = driver.page_source
    df = pd.read_html(html)[0]
    csv = '../data/vote/{}/{}_{}.csv'.format(sido, ssg_sido, ssg_gusingu)
    df.to_csv(csv, index=False)
```

```python
for sido, ssg_sido, ssg_gusigun in req:
    get_data(sido, ssg_sido, ssg_gusigun)
    time.sleep(1)
```
