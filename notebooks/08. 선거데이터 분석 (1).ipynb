```python
# 색상 부여
from matplotlib.colors import LinearSegmentedColormap

color = {
    '일치':'#000000',
    '불일치':'#8f8f8f',
}
cm = LinearSegmentedColormap.from_list('election', list(color.values()), N=2)

plt.figure(figsize=(20, 20))

plt.pcolor(plot_map, cmap=cm, edgecolor='#aaaaaa', linewidth=0.5)

for idx, row in actual.iterrows():
    name = row['표기이름']
    plt.annotate(name, (row['X']+0.5, row['Y']+0.5), weight='bold',
                 fontsize=15, ha='center', va='center', color='white')

for path in BORDER_LINES:
    x, y = path
    plt.plot(x, y, c='black', lw=3)
    
plt.gca().invert_yaxis()

plt.axis('off')
plt.tight_layout()
plt.savefig('../여론조사 일치 여부.png')
plt.show()
```
